```python
import math
```

```python
def ellipse_coordinates(a, b, angle):
    """
    Calculate the coordinates of points on an ellipse for a given angle.

    Parameters:
        a (float): Semi-major axis of the ellipse.
        b (float): Semi-minor axis of the ellipse.
        angle (float): Angle (in radians) from the positive x-axis.

    Returns:
        tuple: Coordinates (x, y) on the ellipse.
    """
    # Parametric equations of the ellipse
    x = a * math.cos(angle)
    y = b * math.sin(angle)

    return x, y
```

```python
# Ellipse parameters
a = 70  # Semi-major axis (R1 in the diagram)
b = 60  # Semi-minor axis (R2 in the diagram)

# Angles for the given points (in radians)
angles = [math.radians(angle) for angle in range(0, 181, 10)]  # Angles from 0° to 180° in 10° increments

# Calculate coordinates for each angle
coordinates = [ellipse_coordinates(a, b, angle) for angle in angles] # This line is added to calculate and store the coordinates

# Print results
for i, (angle, coords) in enumerate(zip(range(0, 181, 10), coordinates)):
    print(f"Coordinates for angle {angle}°: {coords}")
```
